```python
import numpy as np
import matplotlib.pyplot as plt

mu = 7
sigma = 2

def f(x):
    return (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(x-mu)**2/(2*sigma**2))

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = f(x)

plt.plot(x, y, 'b-', linewidth=2)
plt.axvline(mu, color='k', linestyle='--', alpha=0.5, label=f'μ = {mu}')
plt.title(f'Плотность нормального распределения N({mu}, {sigma}^2)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
