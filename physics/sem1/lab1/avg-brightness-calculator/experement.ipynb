### Информация

```python
print(f"max: {max(data)}\nmin: {min(data)}")
```

```python
mean = np.mean(data)
print(f"Среднее: {mean}")

dispersion = 1/(50-1)*sum([(i-np.mean(data))**2 for i in data])
print(f"Дисперсия: {dispersion}")

std = np.sqrt(dispersion)
print(f"Среднеквадратичное отклонение: {std}")

max_ro = 1/(std*np.sqrt(2*np.pi))
print(f"Макс. значение плотности распределения: {max_ro}")
```

```python
f = lambda x: max_ro*np.exp(-(x-mean)**2/(2*dispersion))

while True:
    print(f(float(input())))
```

```python
for n in [1,2,3]:
    _ = len(list(filter(lambda x: mean - std*n <= x <= mean + std*n, data)))

    print(f"{n} - from {mean - std*n:.2f} to {mean + std*n:.2f}: {_} {_/50:.2f}")
```

```python
mean_std = np.sqrt(1/(50*49)*sum(map(lambda x: (x-mean)**2, data)))

print(mean_std)

trust_range = 2.01 * mean_std

print(trust_range)
```

```python
ttl_pogreshnost = np.sqrt(trust_range**2 + (2/3*0.01)**2)
print(ttl_pogreshnost)
```

```python
print(f"{ttl_pogreshnost/mean*100:.2f}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

# Создаем данные с пропуском в начале
x = np.arange(5, 15)
y = np.random.rand(10) * 100

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

# Верхняя панель (пропуск)
ax1.plot(x, y, 'o-')
ax1.set_ylim(90, 100)  # Показываем только верхнюю часть
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(axis='x', which='both', bottom=False)

# Нижняя панель (основные данные)
ax2.plot(x, y, 'o-')
ax2.set_ylim(0, 20)  # Показываем основную часть

# Добавляем линии разрыва
d = 0.015  # Размер линий разрыва
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # левый разрыв
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # правый разрыв

kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # левый разрыв
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # правый разрыв

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Создаем фигуру с двумя subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6), sharey=True,
                              gridspec_kw={'width_ratios': [1, 4]})

# Первый график (левая часть)
ax1.bar([2.48], [0.08], width=0.05, alpha=0.5, color='skyblue',
        edgecolor='black', linewidth=0.5, align="center")
ax1.scatter([2.48], [0.08], color='orange', s=40, marker='o')

# Второй график (правая часть)
ax2.bar([2.61, 2.74, 2.87, 3.00, 3.13, 3.26],
        [0.05, 1.58, 2.51, 2.01, 0.82, 0.17],
        width=0.05, alpha=0.5, color='skyblue',
        edgecolor='black', linewidth=0.5, align="center")
ax2.scatter([2.61, 2.74, 2.87, 3.00, 3.13, 3.26],
            [0.05, 1.58, 2.51, 2.01, 0.82, 0.17],
            color='orange', s=40, marker='o')

# Нормальное распределение
mean = np.mean([2.48, 2.61, 2.74, 2.87, 3.00, 3.13, 3.26])
std = np.std([2.48, 2.61, 2.74, 2.87, 3.00, 3.13, 3.26])
x = np.linspace(2.6, 3.3, num=100)
ax2.plot(x, stats.norm.pdf(x, mean, std), color="#6263c1")

# Настройки
for ax in [ax1, ax2]:
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.set_ylim(0, 3)  # Настройте по необходимости

# Скрываем правую границу первого и левую границу второго графика
ax1.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)

# Добавляем диагональные линии для обозначения разрыва
d = 0.015  # размер диагональных линий
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((1-d, 1+d), (-d, +d), **kwargs)
ax1.plot((1-d, 1+d), (1-d, 1+d), **kwargs)

kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1-d, 1+d), **kwargs)
ax2.plot((-d, +d), (-d, +d), **kwargs)

# Подписи осей
fig.text(0.5, 0.04, 'Время (t), с', ha='center', fontsize=10)
fig.text(0.04, 0.5, 'Плотность вероятности (ρ), c^-1', va='center', 
         rotation='vertical', fontsize=10)

plt.tight_layout()
plt.subplots_adjust(wspace=0.05)
plt.show()
```
