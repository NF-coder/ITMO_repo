### Информация

```python
print(f"max: {max(data)}\nmin: {min(data)}")
```

```python
mean = np.mean(data)
print(f"Среднее: {mean}")

dispersion = 1/(50-1)*sum([(i-np.mean(data))**2 for i in data])
print(f"Дисперсия: {dispersion}")

std = np.sqrt(dispersion)
print(f"Среднеквадратичное отклонение: {std}")

max_ro = 1/(std*np.sqrt(2*np.pi))
print(f"Макс. значение плотности распределения: {max_ro}")
```

```python
f = lambda x: max_ro*np.exp(-(x-mean)**2/(2*dispersion))

while True:
    print(f(float(input())))
```

```python
for n in [1,2,3]:
    _ = len(list(filter(lambda x: mean - std*n <= x <= mean + std*n, data)))

    print(f"{n} - from {mean - std*n:.2f} to {mean + std*n:.2f}: {_} {_/50:.2f}")
```

```python
mean_std = np.sqrt(1/(50*49)*sum(map(lambda x: (x-mean)**2, data)))

print(mean_std)

trust_range = 2.01 * mean_std

print(trust_range)
```

```python
ttl_pogreshnost = np.sqrt(trust_range**2 + (2/3*0.01)**2)
print(ttl_pogreshnost)
```

```python
print(f"{ttl_pogreshnost/mean*100:.2f}%")
```

```python
ax = plt.axes()
ax.yaxis.set_minor_locator(MultipleLocator(0.25))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))

ax.set_ylim(0, mean+0.25)

# histogram
plt.bar(
    [2.48, 2.61, 2.74, 2.87, 3.00, 3.13, 3.26],
    [0.08, 0.05, 1.58, 2.51, 2.01, 0.82, 0.17],
    width=(max(data)-min(data))/8,
    alpha=0.5, 
    color='skyblue',
    edgecolor='black',
    linewidth=0.5,
    align="center",
    
    label='Гистограмма распределения случайной величины'
)

# normal dist
x = np.linspace(min(data), max(data), num=100)
plt.plot(
    x,
    stats.norm.pdf(x, mean, std),
    color ="#6263c1",
    
    label='Плотность вероятности'
)

# points
plt.scatter(
    [2.48, 2.61, 2.74, 2.87, 3.00, 3.13, 3.26],
    [0.08, 0.05, 1.58, 2.51, 2.01, 0.82, 0.17],
    color='orange',
    s=40,
    marker='o'
)

# plot settings
plt.grid(True, alpha=0.5, linestyle='--', which='both')

plt.ylabel('Плотность вероятности (ρ), c^-1\n ΔN/(NΔt), c^-1', fontsize = 10) # подпись оси x
plt.xlabel('Время (t), с', fontsize = 10) # подпись оси y

plt.legend(
    loc="upper right",
    fontsize=9
)
plt.tight_layout()

plt.savefig("plot.png", dpi=1200)

plt.show()
```
